# Creating the Animation for the Lorentz attractor

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Lo1=np.loadtxt('Data/Lorentz4.csv')
Lo2=np.loadtxt('Data/Lorentz5.csv')
Lo3=np.loadtxt('Data/Lorentz6.csv')
Lo1[1,:]
```

```python
#Lo1=Lo1.T
#L1=Lo1[:,1:4]
L1=Lo1.T
L2=Lo2.T
L3=Lo3.T
np.shape(L1)
```

```python
# Plot the Lorenz attractor using a Matplotlib 3D projection.

ax = plt.figure(figsize=(10,10)).add_subplot(projection='3d')

#ax.scatter(*L1[1:4],c= L1[0], s=1, alpha=0.3)
ax.plot(*L1[1:4], alpha=1, linewidth=.2)
ax.plot(*L1[1:4,9999], 'bo', linewidth=.1, markersize=3)

ax.plot(*L2[1:4], 'r', alpha=0.5, linewidth=.2)
ax.plot(*L2[1:4,9999], 'ro', linewidth=.1, markersize=3)

ax.plot(*L3[1:4], 'k', alpha=0.5, linewidth=.2)
ax.plot(*L3[1:4,9999], 'ko', linewidth=.1, markersize=3)

ax.axes.set_xlim3d(left=-15, right=15) 
ax.axes.set_ylim3d(bottom=-20, top=20) 
ax.axes.set_zlim3d(bottom=10, top=38) 
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")
ax.view_init(20, 20, 0)
ax.figsize=(20,20) 

#plt.grid(False)
#plt.axis('off')
#plt.tight_layout()

plt.show()
```

```python
n=3
l=int(L1.shape[1]/n)
plt_folder='ani3/'

for idx in range(l):
    ## creatingh the plot
    ax = plt.figure().add_subplot(projection='3d')
    
    ## plotting the crvs and their current last location
    ax.plot(*L1[1:4,0:idx*n], alpha=0.3, linewidth=.2)
    ax.plot(*L1[1:4,idx*n], 'bo', linewidth=.1, markersize=3)

    ax.plot(*L2[1:4,0:idx*n], 'r', alpha=0.3, linewidth=.2)
    ax.plot(*L2[1:4,idx*n], 'ro', linewidth=.1, markersize=3)

    ax.plot(*L3[1:4,0:idx*n], 'k', alpha=0.3, linewidth=.2)
    ax.plot(*L3[1:4,idx*n], 'ko', linewidth=.1, markersize=3)    
    
    ## Plot layout
    ax.axes.set_xlim3d(left=-15, right=15) 
    ax.axes.set_ylim3d(bottom=-20, top=20) 
    ax.axes.set_zlim3d(bottom=10, top=38) 
    ax.set_xlabel("X Axis")
    ax.set_ylabel("Y Axis")
    ax.set_zlabel("Z Axis")
    ax.set_title('Lorenz Attractor t='+f'{L1[0,idx*n]:.2f}')
    
    azi=idx/l*360
    ax.view_init(10, azi , 0)
    plt.grid(False)
    plt.axis('off')
    
    ## saving the plots with ascending number
    
    plt_name=plt_folder+'L1_'+f'{idx:05d}'+'.png'
    plt.savefig(plt_name)
    plt.close()
    print('saved frame '+f'{idx}' + 'of '+f'{l-1}')
    #plt.show
```
