# Creating the Animation for the Lorentz attractor

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Lo1=np.loadtxt('Data/Lorentz1.csv')
Lo2=np.loadtxt('Data/Lorentz2.csv')
Lo3=np.loadtxt('Data/Lorentz3.csv')
Lo1[1,:]
```

```python
#Lo1=Lo1.T
#L1=Lo1[:,1:4]
L1=Lo1.T
np.shape(L1)
```

```python
# Plot the Lorenz attractor using a Matplotlib 3D projection.

ax = plt.figure().add_subplot(projection='3d')

#ax.scatter(*L1[1:4],c= L1[0], s=1, alpha=0.3)
ax.plot(*L1[1:4], alpha=1, linewidth=.2)
ax.axes.set_xlim3d(left=-25, right=25) 
ax.axes.set_ylim3d(bottom=-30, top=30) 
ax.axes.set_zlim3d(bottom=0, top=50) 
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```

```python
n=10
l=int(L1.shape[1]/n)

for idx in range(l):
    #print(idx)
    ax = plt.figure().add_subplot(projection='3d')

    ax.plot(*L1[1:4,0:idx*n], alpha=1, linewidth=.2)
    ax.axes.set_xlim3d(left=-25, right=25) 
    ax.axes.set_ylim3d(bottom=-30, top=30) 
    ax.axes.set_zlim3d(bottom=0, top=50) 
    ax.set_xlabel("X Axis")
    ax.set_ylabel("Y Axis")
    ax.set_zlabel("Z Axis")
    ax.set_title('Lorenz Attractor t='+f'{L1[0,idx*n]:.3f}')
    plt_folder='ani/'
    plt_name=plt_folder+'L1_'+f'{idx:05d}'+'.png'
    plt.savefig(plt_name)
    plt.close('saved frame '+f'{idx}' + 'of '+f'{l}')
    print()
    #plt.show
```
